```python
import missingno

print("Missingno Version : {}".format(missingno.__version__))
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
starbucks_locations = pd.read_csv("datasets/directory.csv")

starbucks_locations.head()
```

```python
london_housing = pd.read_csv("datasets/housing_in_london_yearly_variables.csv")

london_housing.head()
```

```python
# this does show the missing values

missingno.bar(london_housing, figsize=(10,5), fontsize=12)
```

```python

missingno.matrix(london_housing,color=(0.5,0,0), figsize=(10,5), fontsize=12)
```

```python
missingno.bar(london_housing, color="dodgerblue", sort="ascending", figsize=(10,5), fontsize=12)
```

```python
# combing the two plots as subplots
fig = plt.figure(figsize=(15,7))

ax1 = fig.add_subplot(1,2,1)
missingno.bar(starbucks_locations, color="tomato", fontsize=12, ax=ax1);

ax2 = fig.add_subplot(1,2,2)
missingno.bar(starbucks_locations, log=True, color="tab:green", fontsize=12, ax=ax2);

plt.tight_layout()
```

```python
missingno.matrix(starbucks_locations,sparkline=False,figsize=(10,5), fontsize=12);
```

```python
missingno.matrix(london_housing,sparkline=False,figsize=(10,5), fontsize=12);
```

```python
# the heatmap shows the correlation between the missing values.
# the darker the color, the more correlated the missing values are
missingno.heatmap(starbucks_locations, figsize=(10,5), fontsize=12);
```

The nullity correlation heatmap shows how strongly the presence or absence of one variable affects the presence of another
# 1 means that if one variable appears then the other definitely will
# -1 means that if one variable appears then the other definitely will not
# 0 means that there is no dependence between the occurrence of the two variables

in the heatmap below, we can see that the area_size and no_of_houses are positively correlated

```python
missingno.heatmap(london_housing, cmap="RdYlGn", figsize=(10,5), fontsize=12);


```

# THE DENDROGRAM
IT CLUSTERS THE VARIABLES THAT HAVE SIMILAR NULLITY CORRELATION TOGETHER

```python
missingno.dendrogram(london_housing, figsize=(10,5), fontsize=12);
```

```python
fig = plt.figure(figsize=(15,7))

ax1 = fig.add_subplot(1,2,1)
missingno.dendrogram(starbucks_locations, orientation="right", method="centroid", fontsize=12, ax=ax1);

ax2 = fig.add_subplot(1,2,2)
missingno.dendrogram(london_housing, orientation="top", method="ward", fontsize=12, ax=ax2);

plt.tight_layout()
```
